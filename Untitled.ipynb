# üéØ Analytical Solution of 1D Attenuated Acoustic Wave Equation

## üìå 1. Introduction
Acoustic waves propagate through various media, but real-world materials introduce **attenuation**, causing a gradual loss of wave energy. The **1D attenuated acoustic wave equation** is:


$ 
\frac{\partial^2 p}{\partial t^2} + \alpha \frac{\partial p}{\partial t} = c^2 \frac{\partial^2 p}{\partial x^2}
$

where:
-  p(x,t)  ‚Üí Acoustic pressure at position \( x \) and time \( t \),
-  c  ‚Üí Speed of sound in the medium,
- $ \alpha \ $‚Üí Attenuation coefficient (controls energy dissipation).

---

## üîç 2. Analytical Solution
For an **initial Gaussian wave packet**, the analytical solution is:

$ 
p(x,t) = e^{-\frac{\alpha}{2} t} \left[ e^{-(x - c t)^2} + e^{-(x + c t)^2} \right]
 $

where:
- $  e^{-\frac{\alpha}{2} t}  $ represents the **exponential decay** due to attenuation.
- $ e^{-(x - c t)^2} $ describes a **right-moving wave**.
- $ e^{-(x + c t)^2} $ describes a **left-moving wave**.

---

## üìå 3. Key Observations
‚úîÔ∏è The **wave propagates symmetrically** in both directions.  
‚úîÔ∏è The **amplitude decreases** over time due to attenuation.  
‚úîÔ∏è The **Gaussian shape is preserved**, meaning the wave does not distort.  
‚úîÔ∏è Higher **attenuation $ \alpha $ leads to faster energy loss**.

---

## üèóÔ∏è 4. Physical Interpretation
- In real-world applications such as **ultrasound imaging and medical acoustics**, waves lose energy due to tissue absorption.
- The **wave propagates**, but its strength diminishes over time.
- The equation helps model **sound propagation in biological tissues**, making it useful for applications like **tumor detection and therapeutic ultrasound**.

---





```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Define constants
c = 1.0       # Speed of sound
alpha = 0.5   # Attenuation coefficient
x0 = 0        # Initial wave center

# Define spatial and temporal domains
x = np.linspace(-5, 5, 200)  # Space grid 200 points
t = np.linspace(0, 2, 50)    # Time grid 50 snapshots

# Create mesh grid for plotting
X, T = np.meshgrid(x, t)

# Compute analytical solution
def analytical_solution(x, t, c=1.0, alpha=0.5):
    return np.exp(-alpha * t / 2) * (np.exp(-(x - c * t)**2) + np.exp(-(x + c * t)**2))

P = analytical_solution(X, T, c, alpha)

# Plot results
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# Contour plot of wave propagation
plt.contourf(X, T, P, levels=50, cmap="jet")
plt.colorbar(label="Acoustic Pressure (p)")
plt.xlabel("Position (x)")
plt.ylabel("Time (t)")
plt.title("Analytical Solution of 1D Attenuated Acoustic Wave Equation")
plt.show()

# Line plot for different time steps
plt.figure(figsize=(10, 5))
for t_idx in [0, 10, 20, 30, 40]:  # Select time steps
    plt.plot(x, P[t_idx, :], label=f"t = {t[t_idx]:.2f}")

plt.xlabel("Position (x)")
plt.ylabel("Acoustic Pressure (p)")
plt.title("Wave Propagation Over Time")
plt.legend()
plt.grid()
plt.show()
```

## üìå 5. Conclusion
The **1D attenuated acoustic wave equation** models wave behavior in **dissipative media**. The **exponential decay term** ensures that the wave amplitude decreases, while the Gaussian shape remains intact. This is essential in applications like **biomedical imaging, material testing, and geophysics**.

```python
# Set random seed for reproducibility
dde.config.set_random_seed(42)


# Define constants
c = 1.0       # Speed of sound
alpha = 0.5   # Attenuation coefficient


# Define the PDE
def pde(x, p):
    dp_t = dde.grad.jacobian(p, x, i=0, j=1)          # ‚àÇp/‚àÇt
    dp_tt = dde.grad.hessian(p, x, i=0, j=1)           # ‚àÇ¬≤p/‚àÇt¬≤
    dp_xx = dde.grad.hessian(p, x, i=0, j=0)           # ‚àÇ¬≤p/‚àÇx¬≤
    return dp_tt - c**2 * dp_xx + alpha * dp_t


# Analytical solution for comparison
def analytical_solution(x, t, c=1.0, alpha=0.5):
    return np.exp(-alpha * t / 2) * (np.exp(-(x - c * t)**2) + np.exp(-(x + c * t)**2))

##################################
# Initial condition p(x, 0) = exp(-x^2)
#def ic_func(x):
 #   return np.exp(-x**2)
def ic_func(x):
    return 2*np.exp(-x[:, 0:1]**2)  # p(x,0) = exp(-x¬≤)
#def ic_func(x):
    #return np.exp(-(x[:, 0:1] - 1.0)**2) + np.exp(-(x[:, 0:1] + 1.0)**2)  # Two Gaussians at x=-1 and x=1
# Initial time derivative: ‚àÇp/‚àÇt(x, 0)
#def initial_dpdt(x):
    #alpha = 0.5
    #return -alpha * np.exp(-x[:, 0]**2)
####################################

# Geometry and time domain
geom = dde.geometry.Interval(-5, 5)
timedomain = dde.geometry.TimeDomain(0, 2)
geomtime = dde.geometry.GeometryXTime(geom, timedomain)


# Initial condition and boundary condition
bc = dde.DirichletBC(geomtime, lambda x: 0, lambda _, on_boundary: on_boundary)
#bc = dde.NeumannBC(geomtime, lambda x: 0, lambda _, on_boundary: on_boundary)

#ic = dde.IC(geomtime, lambda x: np.zeros((len(x), 1)), lambda _, on_initial: on_initial)
ic = dde.IC(geomtime, ic_func, lambda _, on_initial: on_initial)

#ic_dpdt = dde.OperatorBC(geomtime, initial_dpdt, lambda x, on_initial: on_initial)


# Setting Up Training Data
data = dde.data.TimePDE(
    geomtime,
    pde,
    [ic, bc],
    num_domain=4000,
    num_boundary=100,
    num_initial=100,
    train_distribution="uniform",
    num_test=1000,
)

# Building the Neural Network
    #net = dde.maps.FNN([2] + [50] * 3 + [1], "tanh", "Glorot uniform")
    #Input: 2 (x and t).
    #2 hidden layers with 90 neurons each.
    #Output: 1 (p).
net = dde.maps.FNN([2] + [90] * 2 + [1], "tanh", "Glorot normal")

# PINN model
model = dde.Model(data, net)

# Compile and train the model
model.compile("adam", lr=1e-4)
losshistory, train_state = model.train(epochs=10000)


# Plot loss history (Shows how the error decreases over training.)
dde.utils.plot_loss_history(losshistory)
plt.title("Training Loss History")
plt.show()

# Prediction
x = np.linspace(-5, 5, 200)
t = np.linspace(0, 2, 50)
X, T = np.meshgrid(x, t)
X_flat = np.hstack((X.flatten()[:, None], T.flatten()[:, None]))

p_pred = model.predict(X_flat).reshape(X.shape)
p_true = analytical_solution(X, T, c, alpha)

# Plotting predicted solution
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")
plt.contourf(X, T, p_pred, levels=50, cmap="jet")
plt.colorbar(label="Acoustic Pressure (p)")
plt.xlabel("Position (x)")
plt.ylabel("Time (t)")
plt.title("PINN Solution of 1D Attenuated Acoustic Wave Equation")
plt.show()


# Plotting analytical solution
plt.figure(figsize=(12, 6))
plt.contourf(X, T, p_true, levels=50, cmap="jet")
plt.colorbar(label="Acoustic Pressure (p)")
plt.xlabel("Position (x)")
plt.ylabel("Time (t)")
plt.title("Analytical Solution of 1D Attenuated Acoustic Wave Equation")
plt.show()

# Comparison plot for specific time steps
plt.figure(figsize=(10, 6))
for t_idx in [0, 10, 20, 30, 40]:
    plt.plot(x, p_pred[t_idx, :], label=f"Predicted (t = {t[t_idx]:.2f})", linestyle="--")
    plt.plot(x, p_true[t_idx, :], label=f"Analytical (t = {t[t_idx]:.2f})", linestyle="-")
plt.xlabel("Position (x)")
plt.ylabel("Acoustic Pressure (p)")
plt.title("Comparison of PINN and Analytical Solutions at Different Time Steps")
plt.legend()
plt.grid(True)
plt.show()
```
